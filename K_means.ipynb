```python
import pandas as np
import numpy as np1 
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn import svm, datasets, metrics
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = np.read_csv("education.csv")
df_1 = np.read_csv("educnorma.csv",index_col=0)
```

```python
df_1.drop(["performance_online"],axis=1,inplace=True)
```

```python
df_1.drop(["Satisfaction_online_Average","Satisfaction_online_Bad", "Satisfaction_online_Good"],axis=1,inplace=True)
```

```python
# clustering
```

```python
X = df_1
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 50):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 50), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# valeurs déjà vérifiées

init_centers=np1.array([[0]*29,[1]*29,[2]*29, [3]*29,[4]*29,[5]*29,[6]*29,[7]*29])
kmeans = KMeans(n_clusters=8,init=init_centers)
pred_y = kmeans.fit_predict(X)
plt.scatter(X['Internet'], X['Online_mode'], c=pred_y, s=50, cmap='copper')
plt.scatter(kmeans.cluster_centers_[:, 3], kmeans.cluster_centers_[:, 8], s=200, c='#DF88B7',alpha=0.7)
plt.title('Cluster')
plt.xlabel('Internet')
plt.ylabel('Doubts_online')
plt.show()
wcss.append(kmeans.inertia_)
```
